```python
def plot_sales_over_time(df, time_freq='D'):
    """
    Plots the time series of sales volume without the range slider.

    Parameters:
    - df: pandas DataFrame, your dataset containing at least 'InvoiceDate', 'Quantity', and 'UnitPrice'.
    - time_freq: str, the frequency for resampling the time series ('D' for daily, 'W' for weekly, 'M' for monthly, etc.).

    Returns:
    - Plotly graph object of the time series plot.
    """
    # Ensure 'InvoiceDate' is a datetime type
    df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

    # Calculate sales volume
    df['SalesVolume'] = df['Quantity'] * df['UnitPrice']

    # Resample and sum sales volume over the specified time frequency
    sales_over_time = df.resample(time_freq, on='InvoiceDate')['SalesVolume'].sum()
    # Create the time series plot
    fig = px.line(sales_over_time, title='Sales Volume Over Time',
                  labels={'value': 'Sales Volume', 'InvoiceDate': 'Date'},
                  markers=True)  # Adding markers can help visualize individual data points

    # Update the layout to remove the range slider
    fig.update_layout(xaxis_title='Date', yaxis_title='Sales Volume',
                      xaxis=dict(rangeslider=dict(visible=False), type='date'))

    return fig
```

```python
fig = plot_sales_over_time(df2, time_freq='M')  # Monthly sales volume
fig.show()
```

```python
fig = plot_sales_over_time(df2, time_freq='W')  # Weekly sales volume
fig.show()
```

## CHAIN RULES

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df2
```

```python
# Germany Data Frame
df_germany = df2[df2["Country"] == "Germany"]
df_germany.shape
```

```python
df_germany.info()
```

```python
# Product-Invoice Pivot Table
def create_invoice_product_df(dataframe, id=False):
    if id:
        return dataframe.groupby(['InvoiceNo', "StockCode"])['Quantity'].sum().unstack().fillna(0). \
            applymap(lambda x: 1 if x > 0 else 0)
    else:
        return dataframe.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().fillna(0). \
            applymap(lambda x: 1 if x > 0 else 0)
```

```python
# Product-Invoice Matrix for Germany (Using Stock Codes)
gr_inv_pro_df = create_invoice_product_df(df_germany, id=True)
gr_inv_pro_df
```

```python
# Function to Retrieve Product Name by Stock Code
def check_id(dataframe, stockcode):
    product_name = dataframe[dataframe["StockCode"] == stockcode]["Description"].unique()[0]
    return stockcode, product_name
```

```python
# Retrieve Product Name for a Given Stock Code from German Sales Data
check_id(df_germany, '22809')
```

```python
# Generating Frequent Itemsets Using the Apriori Algorithm
frequent_itemsets = apriori(gr_inv_pro_df, min_support=0.01, use_colnames=True)
```

```python
# Identifying Frequent Itemsets with Apriori Algorithm
frequent_itemsets = apriori(gr_inv_pro_df, min_support=0.01, use_colnames=True)
```

```python
# Extracting and Displaying Top Association Rules by Support
rules = association_rules(frequent_itemsets, metric="support", min_threshold=0.01)

rules.sort_values("support", ascending=False).head(5)
```

```python
# Sorting Association Rules by Lift Value
sorted_rules = rules.sort_values("lift", ascending=False)
sorted_rules
```

```python
# Generating Product Recommendations Based on Lift Values
product_id = '22492'
recommendation_list = []


for idx, product in enumerate(sorted_rules["antecedents"]):
    for j in list(product):
        if j == product_id:
            recommendation_list.append(list(sorted_rules.iloc[idx]["consequents"])[0])
            recommendation_list = list( dict.fromkeys(recommendation_list) )
```

```python
# Retrieve Product Name for Stock Code '22492' from German Sales Data
check_id(df_germany, '22492')
```

```python
# Top 5 Recommended Products Based on Association Rules
list_top5 = recommendation_list[0:5]
list_top5
```

```python
# Display Product Names for Top 5 Recommendations
for elem in list_top5:
    print(check_id(df_germany,elem))
```
